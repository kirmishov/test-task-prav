```python
def get_leads_by_source(df, event, source):
    return len(df.loc[(df.event == event) & (~df.entity_id.isin(leads_absorbed)) 
                            & (df.role == source)].entity_id.unique())

# leads_Pravoved = len(all_data.loc[(all_data.event == 'val_new') & (~all_data.entity_id.isin(leads_absorbed)) 
#               & (all_data.role == 'internal')].entity_id.unique())
# leads_CPA = len(all_data.loc[(all_data.event == 'val_new') & (~all_data.entity_id.isin(leads_absorbed)) 
#               & (all_data.role == 'cpa')].entity_id.unique())
# leads_Sat = len(all_data.loc[(all_data.event == 'val_new') & (~all_data.entity_id.isin(leads_absorbed)) 
#               & (all_data.role == 'sat')].entity_id.unique())

# напишем функцию, чтобы не повторяться (выше)
```

```python
# leads_Pravoved, leads_CPA, leads_Sat
```

```python
# аналогично напишем функцию по регионам
def get_leads_by_regions(df, event, regions=None):
    if regions == None:
        return len(df.loc[(df.event == event) 
            & (~df.entity_id.isin(leads_absorbed))].entity_id.unique())
    if regions == 'Other':
        return len(df.loc[(df.event == event) & (~df.entity_id.isin(leads_absorbed)) 
            & (~df.region_id.isin(region_ids['Moscow_MO'] + region_ids['Spb_LO']))].entity_id.unique())
    return len(df.loc[(df.event == event) & (~df.entity_id.isin(leads_absorbed)) 
        & (df.region_id.isin(region_ids[regions]))].entity_id.unique())
```

```python
get_leads_by_regions(all_data, 'val_new', 'Other')
get_leads_by_regions(all_data, 'val_new', 'Moscow_MO')
get_leads_by_regions(all_data, 'val_new', 'Spb_LO')
get_leads_by_regions(all_data, 'val_new')
```
